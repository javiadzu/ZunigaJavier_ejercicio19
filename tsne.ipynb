```python
#Importamos las librerías necesarias
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import sklearn.manifold
import sklearn.datasets
import glob
import sklearn.cluster

#Cargamos el archivo a utilizar
X = pd.read_csv('traini.csv')
```

```python
#Definimos el número de vecinos cercanos con los que se realiza TSNE
tsne = sklearn.manifold.TSNE(perplexity=5)
```

```python
#Utilizamos los datos del archivo
tsne.fit(X)
```

```python
#Lo proyectamos en el espacio 2D
embedding = tsne.embedding_
```

```python
#graficamos los datos
plt.scatter(embedding[:,0], embedding[:,1])
plt.title('T-SNE con perplexity = 5')
```

Puede verse una tendencia a asociarse los datos, o por lo menos forman una linea continua, lo que yo interpreto como pequeñas variaciones entre ellos.

```python
perp =[3,5,8,12,15,18]
x1 = sklearn.manifold.TSNE(perplexity=perp[0])
x2 = sklearn.manifold.TSNE(perplexity=perp[1])
x3 = sklearn.manifold.TSNE(perplexity=perp[2])
x4 = sklearn.manifold.TSNE(perplexity=perp[3])
x5 = sklearn.manifold.TSNE(perplexity=perp[4])
x6 = sklearn.manifold.TSNE(perplexity=perp[5])
x1.fit(X)
x2.fit(X)
x3.fit(X)
x4.fit(X)
x5.fit(X)
x6.fit(X)
```

```python
embedding1 = x1.embedding_
embedding2 = x2.embedding_
embedding3 = x3.embedding_
embedding4 = x4.embedding_
embedding5 = x5.embedding_
embedding6 = x6.embedding_
```

```python
fig, axs = plt.subplots(2, 3, figsize=(15,15))
axs[0, 0].scatter(embedding1[:,0], embedding1[:,1])
axs[0, 0].set_title('perplexity='+str(perp[0]))
axs[0, 1].scatter(embedding2[:,0], embedding3[:,1])
axs[0, 1].set_title('perplexity='+str(perp[1]))
axs[0, 2].scatter(embedding3[:,0], embedding3[:,1])
axs[0, 2].set_title('perplexity='+str(perp[2]))
axs[1, 0].scatter(embedding4[:,0], embedding4[:,1])
axs[1, 0].set_title('perplexity='+str(perp[3]))
axs[1, 1].scatter(embedding5[:,0], embedding3[:,1])
axs[1, 1].set_title('perplexity='+str(perp[4]))
axs[1, 2].scatter(embedding6[:,0], embedding3[:,1])
axs[1, 2].set_title('perplexity='+str(perp[5]))
```

A medida que aumenta la perplexity los datos empiezan a alejarse y agruparse en grupos más localizados. En definitiva
